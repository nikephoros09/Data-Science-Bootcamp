```python
import pandas as pd
import sqlite3 
import matplotlib.pyplot as plt
connection = sqlite3.connect("../../data/checking-logs.sqlite")
```

```python
##WORKING DAYS
query = '''
SELECT
    COUNT (*) AS Working_days
FROM checker
WHERE uid LIKE 'user%'
  AND strftime('%w', timestamp) IN ('1','2','3','4','5');
'''
pd.read_sql(query, connection)
```

```python
##WEEKEND
query = '''
SELECT
    COUNT (*) AS Weekend
FROM checker
WHERE uid LIKE 'user%'
  AND strftime('%w', timestamp) IN ('0','6');
'''
pd.read_sql(query, connection)
```
