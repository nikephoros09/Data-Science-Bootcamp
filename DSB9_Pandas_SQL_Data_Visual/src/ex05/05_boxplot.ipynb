```python
import pandas as pd 
import sqlite3
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../../data/ab-test.csv")
after_ctrl = df[(df['time'] == 'after') & (df['group'] == 'control')]
before_ctrl = df[(df['time'] == 'before') & (df['group'] == 'control')]
after_test = df[(df['time'] == 'after') & (df['group'] == 'test')]
before_test = df[(df['time'] == 'before') & (df['group'] == 'test')]
```

```python
after_ctrl.info()
before_ctrl.info()
after_test.info()
before_test.info()
```

```python
figure = plt.figure(figsize=(15, 8))

ax_left  = figure.add_subplot(1, 2, 1)
ax_right = figure.add_subplot(1, 2, 2, sharey=ax_left)

ax_left.boxplot(
    [after_ctrl['diff'], after_test['diff']],
    tick_labels=['contol', 'test'],
    boxprops={'color': 'green', 'linewidth': 3},
    medianprops={'color': 'red', 'linewidth': 2},
    whiskerprops={'color': 'red'},
    capprops={'color': 'red'},
    showmeans=True
)
ax_left.grid()

ax_right.boxplot(
    [before_ctrl['diff'], before_test['diff']],
    tick_labels=['contol', 'test'],
    boxprops={'color': 'green', 'linewidth': 3},
    medianprops={'color': 'red', 'linewidth': 2},
    whiskerprops={'color': 'red'},
    capprops={'color': 'red'},
    showmeans=True
)
ax_right.grid()

figure.suptitle('Lag before labs deadlines. After and before Newsfeed', fontsize=15)
plt.show()
```
