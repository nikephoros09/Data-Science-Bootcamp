## create a dataframe views with two columns

```python
import pandas as pd
```

```python
data = pd.read_csv(
    "../../data/feed-views.log",
    sep="\t",
    header=None,
    names=["datetime", "user"]
)
data["datetime"] = pd.to_datetime(data["datetime"])

data["year"] = data["datetime"].dt.year
data["month"] = data["datetime"].dt.month
data["day"] = data["datetime"].dt.day
data["hour"] = data["datetime"].dt.hour
data["minute"] = data["datetime"].dt.minute
data["second"] = data["datetime"].dt.second
```

```python
data.info()
```

## create the new column daytime

```python
bins = [0, 4, 7, 11, 17, 20, 24]
labels = ['night', 'early morning', 'morning', 'afternoon', 'early evening', 'evening']

data['daytime'] = pd.cut(data['hour'], bins=bins, labels=labels, right=False)

data.set_index('user', inplace=True)
```

```python
data.count()
```

## sort values in your dataframe

```python
data_sorted = data.sort_values(by=['hour', 'minute', 'second'], ascending=[True, True, True])
```

## calculate the minimum and maximum

```python
night_mask = data['daytime'] == 'night'
max_hour_night = data.loc[night_mask, 'hour'].max()
```

```python
max_hour_night
```

```python
morning_mask = data['daytime'] == 'morning'
min_hour_morning = data.loc[morning_mask, 'hour'].min()
```

```python
data.loc[
    (data['daytime'] == 'night') & (data['hour'] == max_hour_night)
].head(1)
```
