## download and read the CSV file and make ID the index column

```python
import pandas as pd
df = pd.read_csv('../../data/auto.csv', index_col='ID')
```

## count the number of observations using the method count()

```python
df.count()
```

## drop the duplicates

```python
df.drop_duplicates(
    subset=['CarNumber', 'Make_n_model', 'Fines'],
    keep='last',
    inplace=True
)
```

```python
df.count()
```

## work with missing values

```python
df.isna().sum()
```

```python
df = df.dropna(axis=1, thresh=len(df) - 500)
```

```python
df.isna().sum()
```

```python
#df['Refund'] = df['Refund'].fillna(method='ffill')
df['Refund'] = df['Refund'].ffill()
```

```python
df.isna().sum()
```

```python
df['Fines'] = df['Fines'].fillna(value=df['Fines'].mean())
```

```python
df.isna().sum()
```

## split and parse the make and model 

```python
def parse_mm(value):
    parts = value.split(' ', 1)
    make = parts[0]
    model = parts[1] if len(parts) > 1 else None
    return pd.Series({'Make': make, 'Model': model})

df[['Make', 'Model']] = df['Make_n_model'].apply(parse_mm)
```

```python
df.drop(inplace=True, axis=1, labels='Make_n_model')
```

```python
df.to_json(path_or_buf="../../data/auto.json", orient='records')
```
