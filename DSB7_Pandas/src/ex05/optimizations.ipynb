## read the fines.csv that you saved

```python
import pandas as pd
import gc
df = pd.read_csv('../../data/fines.csv')
df
```

```python
%%timeit
calculated = []
for i in range(0, len(df)):
    res = df.iloc[i]['Fines'] / df.iloc[i]['Refund'] * df.iloc[i]['Year']
    calculated.append(res)
df['Calculated'] = calculated
```

```python
df
```

```python
%%timeit
calculated = []
for _, row in df.iterrows():
    res = row['Fines'] / row['Refund'] * row['Year']
    calculated.append(res)
df['Calculated'] = calculated
```

```python
df
```

```python
%%timeit
df["Calculated"] = df.apply(lambda r: r["Fines"] * r["Year"] / r["Refund"],axis=1)
```

```python
df
```

```python
%%timeit
df['Calcualted'] = df['Fines'] / df['Refund'] * df['Year']
```

```python
%%timeit
df['Calcualted'] = df['Fines'].values / df['Refund'].values * df['Year'].values
```
