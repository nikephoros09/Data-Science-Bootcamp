## create a connection to the database using the library sqlite3

```python
import pandas as pd
import sqlite3
connection = sqlite3.connect('../../data/checking-logs.sqlite')
pd.io.sql.read_sql("PRAGMA table_info(pageviews);", connection)
```

## get only the first 10 rows of the table pageviews

```python
pd.io.sql.read_sql('SELECT * FROM pageviews LIMIT 10;', connection)
```

## get the subtable using only one query

```python
##only uid and datetime are used
pd.io.sql.read_sql("SELECT uid, datetime FROM pageviews LIMIT 10;", connection)
```

```python
query = '''
SELECT uid, datetime FROM pageviews
WHERE uid LIKE 'user_%'
ORDER BY uid;
'''
pageviews = pd.read_sql(query, connection)
pageviews["datetime"] = pd.to_datetime(pageviews["datetime"])
pageviews = pageviews.set_index('datetime')
```
